```python
from apiclient.discovery import build
from apiclient.errors import HttpError
from oauth2client.tools import argparser
```

```python
DEVELOPER_KEY = "api!"
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"

youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=DEVELOPER_KEY)


search_response = youtube.search().list(
    q="전기자동차전기세",
    part="snippet",
    maxResults=1000
  ).execute()
```

```python
search_response
```

```python
keywords = []
description = []


for search_result in search_response.get("items", []):
    if search_result["id"]["kind"] == "youtube#video":
        keywords.append("%s" % (search_result["snippet"]["title"]))
        
for search_result in search_response.get("items", []):
    if search_result["id"]["kind"] == "youtube#video":
        description.append("%s" % (search_result["snippet"]["description"]))
```

```python
with open('/Users/user/Downloads/EV_keywords_youtube0908.txt', 'w', encoding='utf-8') as f:
    for name in keywords:
        f.write(name+'\n')
```

```python
with open('/Users/user/Downloads/EV_description_youtube0908.txt', 'w', encoding='utf-8') as f:
    for name in description:
        f.write(name+'\n')
```

```python
from wordcloud import WordCloud
from PIL import Image
import numpy as np
```

```python
with open("/Users/user/Downloads/EV_keywords_youtube0908.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        print(line)
```

```python
keyword = ''
for line in lines:
    keyword += line
```

```python
with open("/Users/user/Downloads/EV_description_youtube0908.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        print(line)

description = ''
for line in lines:
    description += line
```

```python
from konlpy.tag import Kkma
from konlpy.tag import Okt
import konlpy.tag
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib 

%matplotlib inline
```

```python
okt = konlpy.tag.Okt()
```

```python
noun_keyword = okt.nouns(keyword)
```

```python
noun_keyword
```

```python
count = Counter(noun_keyword)
most_keyword = count.most_common(10000)
print(most_keyword)
```

```python
noun_description = okt.nouns(description)
```

```python
noun_description
```

```python
count2 = Counter(noun_description)
most_description = count2.most_common(10000)
print(most_description)
```

```python
mask = np.array(Image.open('car4.jpeg'))
```

```python
wordcloud = WordCloud(font_path='AppleGothic',
                      background_color='white',
                      width = 480,
                      height = 480,
                     mask=mask)
wordcloud.generate_from_frequencies(dict(most_keyword))

plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='lanczos')
plt.axis("off")
```

```python
wordcloud = WordCloud(font_path='AppleGothic',
                      background_color='white',
                      width = 480,
                      height = 480,
                     mask=mask)
wordcloud.generate_from_frequencies(dict(most_description))

plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='lanczos')
plt.axis("off")
```
